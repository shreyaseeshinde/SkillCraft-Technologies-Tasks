```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the Dataset
data_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv'
df = pd.read_csv(data_path, skiprows=4)  # Skipping metadata rows if necessary
```

```python
# Step 2: Inspect the Data
print(df.head())  # Display the first few rows
print(df.info())  # Display information about the dataset
```

```python
# Step 3: Filter the Data for a Specific Year (e.g., 2020)
df_2020 = df[['Country Name', '2020']].dropna()
```

```python
# Step 4: Create a Bar Chart of Population by Country

# Sort and plot the top 10 countries by population in 2020
top_10_countries = df_2020.sort_values(by='2020', ascending=False).head(10)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10_countries['Country Name'], top_10_countries['2020'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population in 2020')
plt.title('Top 10 Countries by Population in 2020')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 5: Create a Histogram of Population Distribution Across Countries

# Assuming '2020' column contains population data for each country
plt.figure(figsize=(10, 6))
plt.hist(df_2020['2020'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.title('Distribution of Population Across Countries in 2020')
plt.show()
```
